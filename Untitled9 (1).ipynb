```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
sns.set_theme(style="whitegrid")
```

```python
df= pd.read_csv('/content/Amazon Sale Report.csv')
df.info()
df.head()
```

```python
str_columns = [
    'Order ID', 'Status', 'Fulfilment', 'ship-service-level',
    'Style', 'SKU', 'Category', 'ASIN', 'Courier Status', 'currency',
    'ship-city', 'ship-state', 'ship-postal-code', 'ship-country',
    'promotion-ids', 'B2B'
]

for col in str_columns:
    df[col] = df[col].astype(str)

df.dtypes
```

```python
df.shape[0]
```

```python
df.isnull().sum()
```

```python
df.replace('nan', np.nan, inplace=True)
```

```python
unique_count = df["Courier Status"].nunique()
print(f"Farklı Courier Status sayısı: {unique_count}")
df["Courier Status"].unique()
```

```python
df['Amount'] = df['Amount'].fillna(0)
df['currency'] = df['currency'].fillna("NaN")
```

```python
df['Courier Status']= df['Courier Status'].fillna("Cancelled")
```

```python
columns_to_fill = ["ship-city", "ship-state", "ship-postal-code", "ship-country"]

for col in columns_to_fill:
    moda = df[col].mode()[0]  # Sütunun en sık tekrar eden değeri (modu)
    df[col].fillna(moda)  # NaN değerleri mod ile doldurur
```

```python
# Diğer sütunlardaki NaN değerleri ile olan ilişkisini görmek
nan_rows = df[df['promotion-ids'].isna()]  # "promotion-ids" NaN olan satırlar
nan_rows.describe()  # NaN değerlerin olduğu satırların özet istatistikleri
nan_rows.head(10)
```

"Promotion ids" sütunu, satın almanın herhangi bir indirim alıp almadığını belirtir. Bu durumda, eksik (NaN) değerlerin indirim uygulanmayan satın alımları temsil ettiği varsayılmaktadır. Bu nedenle, eksik değerler "N/A" (geçerli değil) ile değiştirilecektir.

```python
df['promotion-ids'] = df['promotion-ids'].fillna("N/A")
```

```python
# Diğer sütunlardaki NaN değerleri ile olan ilişkisini görmek
nan_rows = df[df['fulfilled-by'].isna()]  # "promotion-ids" NaN olan satırlar
nan_rows.describe()  # NaN değerlerin olduğu satırların özet istatistikleri
```

```python
# Courier Status sütunundaki farklı değer sayısı
unique_count = df["fulfilled-by"].nunique()
print(f"Farklı fulfilled-by sayısı: {unique_count}")
df["fulfilled-by"].unique()   # Benzersiz değerleri listeler
```

```python
# Courier Status sütunundaki farklı değer sayısı
unique_count = df["Unnamed: 22"].nunique()
print(f"Farklı Unnamed: 22 sayısı: {unique_count}")
df["Unnamed: 22"].unique()   # Benzersiz değerleri listeler
```

```python
category_revenue = df.groupby('Category')['Amount'].sum()
total_revenue = df['Amount'].sum()
category_revenue_percentage = (category_revenue / total_revenue) * 100
print("Percentage Revenue Per Product Category:")
print(category_revenue_percentage)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=category_revenue_percentage.values, y=category_revenue_percentage.index, hue=category_revenue_percentage.index, palette="viridis", legend=False)
plt.title("Percentage of Product Category for Net Revenue")
plt.xlabel("Percentage Revenue (%)")
plt.ylabel("Product Category")
plt.show()
```

```python
average_purchase_price = df.groupby('Category')['Amount'].mean()
print("Average Purchase Price per Product Category:")
print(average_purchase_price)
plt.figure(figsize=(10, 6))
sns.barplot(x=average_purchase_price.values, y=average_purchase_price.index, palette="magma")
plt.title("Average Purchase Price per Product Category")
plt.xlabel("Average Purchase Price")
plt.ylabel("Product Category")
plt.show()
```

```python
df['Revenue'] = df['Qty'] * df['Amount']

city_revenue = df.groupby('ship-city')['Revenue'].sum().reset_index()

city_revenue = city_revenue.sort_values(by='Revenue', ascending=False)
top_10_cities = city_revenue.sort_values(by='Revenue', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_10_cities['ship-city'], top_10_cities['Revenue'], color='royalblue')
plt.title('Top 10 Cities by Revenue Generated', fontsize=14)
plt.xlabel('ship-city', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
top_cities = df.groupby('ship-city')['Revenue'].sum().reset_index().sort_values(by='Revenue', ascending=False).head(10)
top_city_data = df[df['ship-city'].isin(top_cities['ship-city'])]
category_sales = top_city_data.groupby(['ship-city', 'Category'])['Revenue'].sum().reset_index()
category_pivot = category_sales.pivot(index='ship-city', columns='Category', values='Revenue').fillna(0)

category_pivot.plot(kind='bar', stacked=True, figsize=(12, 6), colormap="tab10")

plt.title("Category Sales Across Top 10 Cities", fontsize=14)
plt.xlabel("ship-city", fontsize=12)
plt.ylabel("Total Revenue ($)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title="Category", bbox_to_anchor=(1,1))
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
